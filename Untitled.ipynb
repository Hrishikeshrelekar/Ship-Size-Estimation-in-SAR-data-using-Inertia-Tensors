```python
def binary_mask(image): 
    ''' Masks out sea clutter using one parametric CFAR algorithm. Pixels will be converted into binary values 0 or 1. 
    Value 1 will be assigned to ship pixels.
    '''
    n=(image.shape[0]*image.shape[1])
    im=np.zeros(a.shape)
    for e in range(len(im)):
        for f in range(im.shape[1]):
            mean=(image.sum()-image[e,f])/(n-1)
            T=((10**(3/n)-1)*n*mean)
            if image[e,f]>T:
                im[e,f]=1
    
    if im.sum()==0:
        print('Error')
    return(im,T)
```

```python
def Inertia_tesnor_eigen(im,x,y):
    '''
    Calculating Inertia tensors of binary image
    '''
    ixx=0
    for i in range(im.shape[0]):
        ixx=ixx+np.sum(im[i,:])*((i-y)**2)
        
    iyy=0
    for i in range(im.shape[1]):
        iyy=iyy+np.sum(im[:,i])*((i-x)**2)
        
    ixy=0
    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            ixy=ixy+im[i,j]*(i-y)*(j-x)
            
    I=np.array([[ixx,-ixy],[-ixy,iyy]])
    w,v=eigh(I)
    m2=v[1,1]/v[0,1]
    m1=v[1,0]/v[0,0] 

    return m1,m2
    
    
def d_matrix(im,x,y,m1,m2):
    '''
    Calculating D matrix of binary image
    '''    

    d1=np.zeros(im.shape)
    d2=np.zeros(im.shape)
    
    if m1==0:
        for i in range(im.shape[0]):
            for j in range(im.shape[1]):
                d1[i,j]=i-y
                d2[i,j]=j-x
                
    elif m2==0:
        for i in range(im.shape[0]):
            for j in range(im.shape[1]):
                d1[i,j]=j-x
                d2[i,j]=i-y
        
    else:
        for i in range(im.shape[0]):
            for j in range(im.shape[1]):
                d1[i,j]=(i-(y+m1*(j-x)))/(np.sqrt(1+m1*m1))
                d2[i,j]=(i-(y+m2*(j-x)))/(np.sqrt(1+m2*m2))
            
    return d1, d2

def rms_distances(im,d1,d2):
    '''
    Calculating RMS distances of binary image
    '''
    s=np.sqrt(np.multiply(np.square(d1[d1<0]),im[d1<0]).sum()/im[d1<0].sum())
    p=np.sqrt(np.multiply(np.square(d1[d1>0]),im[d1>0]).sum()/im[d1>0].sum())
    f=np.sqrt(np.multiply(np.square(d2[d2<0]),im[d2<0]).sum()/im[d2<0].sum())
    a=np.sqrt(np.multiply(np.square(d2[d2>0]),im[d2>0]).sum()/im[d2>0].sum())
              
    return s,p,f,a

def significance(original_im,binary_im):
    '''
    Calculating Significance of an image
    '''    
    mean=original_im[binary_im==0].mean()
    std=original_im[binary_im==0].std()
    return (original_im.max()-mean)/std
            
def significance_mean(original_im,binary_im):
    '''
    Calculating Significance mean of an image
    '''
    mean=original_im[binary_im==0].mean()
    std=original_im[binary_im==0].std()
    return (original_im.mean()-mean)/std


def Hu_moments(original_im):
    '''
    Calculating Hu moments
    '''
    return(moments_hu(original_im))

    
```

```python
for n in range(len(z)):
    a=tif.imread(z[n])[:,:,0] 
    im,thresh=binary_mask(np.copy(a))
    for w in range(20):
        cen_row,cen_col=ndimage.measurements.center_of_mass(np.copy(im))
        m1,m2=Inertia_tesnor_eigen(np.copy(im),cen_col,cen_row)
        if m1==0:
            break
        elif m2==0:
            break
        else:
            d1,d2=d_matrix(np.copy(im),cen_col,cen_row,m1,m2)
            sigma_s,sigma_p,sigma_f,sigma_a=rms_distances(np.copy(im),d1,d2)
            x2=cen_col+2.07*sigma_s/np.sqrt(1+m2*m2)
            y2=cen_row+m2*(x2-cen_col)
            x3=cen_col-2.07*sigma_p/np.sqrt(1+m2*m2)
            y3=cen_row+m2*(x3-cen_col)

            x4=cen_col-2.07*sigma_f/np.sqrt(1+m1*m1)
            y4=cen_row+m1*(x4-cen_col)

            x5=cen_col+2.07*sigma_a/np.sqrt(1+m1*m1)
            y5=cen_row+m1*(x5-cen_col)

            ri=np.sign(y2+m1*(cen_col-x2)-cen_row)
            le=np.sign(y3+m1*(cen_col-x3)-cen_row)
            to=np.sign(y4+m2*(cen_col-x4)-cen_row)
            bo=np.sign(y5+m2*(cen_col-x5)-cen_row)
            ini=len(im[im==0])
            for i in range(im.shape[0]):
                for j in range(im.shape[1]):
                    if np.sign(y2+m1*(j-x2)-i)!=ri:
                        im[i,j]=0
                    if np.sign(y3+m1*(j-x3)-i)!=le:
                        im[i,j]=0
                    if np.sign(y4+m2*(j-x4)-i)!=to:
                        im[i,j]=0
                    if np.sign(y5+m2*(j-x5)-i)!=bo:
                        im[i,j]=0

            if len(im[im==0])==ini:
                break
    
    e.loc[int(z[n].split('\\')[1].split('.')[0]), 'Index'] = int(z[n].split('\\')[1].split('.')[0])
    e.loc[int(z[n].split('\\')[1].split('.')[0]), 'Length'] = 2.07 *(sigma_f+sigma_a)
    e.loc[int(z[n].split('\\')[1].split('.')[0]), 'Width'] = 2.07 *(sigma_s+sigma_p)
    e.loc[int(z[n].split('\\')[1].split('.')[0]),'Significance']=significance(a,im)
    e.loc[int(z[n].split('\\')[1].split('.')[0]),'Total Pixels']=len(a[im==1])
    e.loc[int(z[n].split('\\')[1].split('.')[0]),'Mass']=a[im==1].sum()
    e.loc[int(z[n].split('\\')[1].split('.')[0]),'Area']=2.07*2.07*(sigma_s+sigma_p)*(sigma_f+sigma_a)
    e.loc[int(z[n].split('\\')[1].split('.')[0]),'Significance_mean']=significance_mean(a,im)
    
    hu=Hu_moments(a)
    

    e.loc[int(z[n].split('\\')[1].split('.')[0]),'Hu_1']=hu[0]
    e.loc[int(z[n].split('\\')[1].split('.')[0]),'Hu_2']=hu[1]
    e.loc[int(z[n].split('\\')[1].split('.')[0]),'Hu_3']=hu[2]
    e.loc[int(z[n].split('\\')[1].split('.')[0]),'Hu_4']=hu[3]
    e.loc[int(z[n].split('\\')[1].split('.')[0]),'Hu_5']=hu[4]
    e.loc[int(z[n].split('\\')[1].split('.')[0]),'Hu_6']=hu[5]
    e.loc[int(z[n].split('\\')[1].split('.')[0]),'Hu_7']=hu[6]
    
    
        
    
```
